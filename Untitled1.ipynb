```python
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db=client.airbnb
booking=db.booking
homes_col=db.airbnb_homes
prices_col=db.prices
homes=homes_col.find({})
```

```python
homes=homes_col.find({})
wards=[]
prices={}
count={}
ave={}
for home in homes:
    ward=home["details"]["ward"]+' '+home["details"]["district"]+' '+home["details"]["province"]
    
    if ward not in wards:
        wards.append(ward)
        prices[ward]=0.0
        count[ward]=0
    try:
        prices[ward]+=home["details"]["price"]
    except:
        if count[ward]!=0:
            prices[ward]+=prices[ward]/count[ward]            
    count[ward]+=1
# print(prices)
# print(count)
for key in prices:
    if count[key]!=0:
        ave[key]=prices[key]/count[key]
    else:
        ave[key]=0
print(len(ave.items()))
```
